# Storage formats for weather observation data

```python
import io
import math
import warnings

import fiona
import geopandas as gpd
import numpy as np
import pandas as pd
from attr import attrib, attrs

# https://github.com/geopandas/geopandas/issues/1213
warnings.simplefilter(action="ignore", category=FutureWarning)

WGS84 = "epsg:4326"
NZTM = "epsg:2193"
```

## GeoJSON parser from weatherwatch
Copied from https://github.com/mje-nz/weatherwatch/blob/78681ff440cfcb34bf33f13502db0c69651ad735/weatherwatch.py

```python
def format_wind_dir(angle):
    if np.isnan(angle):
        return ""
    directions = [
        "N", "NNE", "NE", "ENE", "E", "ESE", "SE", "SSE", "S", "SSW", "SW", "WSW", "W",
        "WNW", "NW", "NNW",
    ]
    return directions[math.floor(angle / 22.5 + 0.5) % 16]

@attrs
class ObservationCollection:

    """A collection of WeatherWatch live observation data.
    
    Notes:
        * precipTotal is rain today in mm
        * windSpeed is in km/h
        * windDir is degrees clockwise of north
        * humidity is in %
        * uv seems to be UV index
        * pressure is in hPa
        * realtimeFrequency is currently unused
        * qcStatus is -1 for inaccurate
        * solarRadiation is in W/m2
    """

    obs: gpd.GeoDataFrame = attrib()
    fetch_time: pd.Timestamp = attrib(factory=lambda: pd.Timestamp.utcnow())

    @property
    def age(self):
        """The time since these observations were fetched."""
        return pd.Timestamp.utcnow() - self.fetch_time

    @classmethod
    def from_geojson(cls, data):
        """Load WeatherWatch observations into a GeoDataFrame."""
        with fiona.open(io.BytesIO(data)) as features:
            obs = gpd.GeoDataFrame.from_features(features, crs=features.crs)
        assert sorted(obs.columns) == [
            "dewpt", "geometry", "humidity", "major", "neighborhood", "obsTimeUtc",
            "precipTotal", "pressure", "qcStatus", "realtimeFrequency", "softwareType",
            "solarRadiation", "stationID", "temp", "uv", "windSpeed", "winddir",
        ], f"Unexpected columns: {list(obs.columns)}"

        # Project to NZTM
        obs.to_crs(NZTM, inplace=True)

        # Set column dtypes
        numeric_columns = [
            "temp", "humidity", "dewpt", "uv", "solarRadiation", "precipTotal",
            "pressure", "windSpeed", "winddir", "qcStatus"
        ]
        obs[numeric_columns] = obs[numeric_columns].apply(pd.to_numeric, errors="coerce")
        obs["obsTimeUtc"] = pd.to_datetime(obs.obsTimeUtc, utc=True)

        # Tidy temperatures
        invalid_temps = (obs.temp < -25) | (obs.temp > 40)
        obs["temp"] = obs.temp.mask(invalid_temps)
        obs["dewpt"] = obs.dewpt.mask(invalid_temps)

        # Tidy humidities
        invalid_humidities = (obs.humidity < 0) | (obs.humidity > 100)
        obs["humidity"] = obs.humidity.mask(invalid_humidities)
        obs["dewpt"] = obs.dewpt.mask(invalid_humidities)

        # Tidy UVs
        invalid_uvs = (obs.uv < 0) | (obs.uv > 20)
        obs["uv"] = obs.uv.mask(invalid_uvs)

        # Tidy solar radiations
        invalid_solar_rads = (obs.solarRadiation < 0) | (obs.solarRadiation > 1500)
        obs["solarRadiation"] = obs.solarRadiation.mask(invalid_solar_rads)

        # Tidy rainfalls
        invalid_rains = (obs.precipTotal < 0) | (obs.precipTotal > 1000)
        obs["precipTotal"] = obs.precipTotal.mask(invalid_rains)

        # Tidy pressures
        invalid_pressures = (obs.pressure < 900) | (obs.pressure > 1080)
        obs["pressure"] = obs.pressure.mask(invalid_pressures)

        # Tidy wind speeds
        invalid_wind_speeds = (obs.windSpeed < 0) | (obs.windSpeed > 250)
        obs["windSpeed"] = obs.windSpeed.mask(invalid_wind_speeds)

        # Tidy wind dirs
        obs['windDirRaw'] = obs.winddir
        invalid_wind_dirs = (obs.windDirRaw < 0) | (obs.windDirRaw > 360)
        obs['windDirRaw'] = obs.windDirRaw.mask(invalid_wind_dirs)
        obs['windDir'] = obs.windDirRaw.apply(format_wind_dir)

        # Reorder and rename columns
        obs.rename(columns={"dewpt": "dewPt", "stationID": "stationId"}, inplace=True)
        obs = obs[[
            "stationId", "temp", "humidity", "dewPt", "uv", "solarRadiation",
            "precipTotal", "pressure", "windSpeed", "windDir", "windDirRaw",
            "obsTimeUtc", "qcStatus", "neighborhood", "geometry"
        ]]
        obs.set_index("stationId", inplace=True)

        return cls(obs)
```

```python
from tqdm.notebook import tqdm as tqdm_base
def tqdm(*args, **kwargs):
    """tqdm progress which shows current value."""
    with tqdm_base(*args, **kwargs) as t:
        for i in t:
            # Hack to show something sensible
            current = i
            if type(i) in (tuple, list):
                current = i[0]
            t.set_postfix({"current": current})
            yield i
        t.set_postfix({})
```

```python
def gdf_to_df(gdf):
    """Convert a GeoDataFrame to a DataFrame."""
    gdf = gdf.copy()
    gdf["x"], gdf["y"] = gdf.geometry.x, gdf.geometry.y
    return pd.DataFrame(gdf.drop(columns="geometry"))
```

## Set up benchmarks

```python
def _measure(filename, writer, reader, verbose=False):
    """Run writer and reader until they've both taken longer than a
    second, then return file size lists of read/write times."""
    try:
        gc.disable()
        file_sizes = []
        read_durations = []
        write_durations = []
        for i in range(1, 1000):
            start_time = time.time()
            writer(filename)
            write_durations.append(time.time() - start_time)
            
            start_time = time.time()
            reader(filename)
            read_durations.append(time.time() - start_time)
            
            file_sizes.append(os.path.getsize(filename))
            os.remove(filename)
            if (np.sum(read_durations) > 1 and np.sum(write_durations) > 1) \
                    or np.sum(read_durations + write_durations) > 5:
                break
        return file_sizes, write_durations, read_durations
    finally:
        gc.enable()
    
    
def measurements_to_df(measurements):
    """Convert a list of (format, compression, file size, write time, read time) tuples
    into a DataFrame."""
    return pd.concat(
        [pd.DataFrame({
            "format": f,
            "compression": c or "",
            "file_size": fs,
            "write_time": w,
            "read_time": r
        }) for (f, c, fs, w, r) in measurements]
    )
```

```python
# GeoJSON
def measure_geojson(raw_data, compression_module=None):
    compressor = decompressor = lambda t: t
    if compression_module is not None:
        compressor = compression_module.compress
        decompressor = compression_module.decompress
    if type(raw_data) in (list, tuple):
        raw_data = b"\0".join(raw_data)
    
    return _measure(
        "tmp.geojson",
        # Writing to GeoJSON is free apart from compression overhead
        lambda f: open(f, "wb").write(compressor(raw_data)),
        # But reading from GeoJSON also includes parsing it
        lambda f: [ObservationCollection.from_geojson(d).obs
                   for d in decompressor(open(f, "rb").read()).split(b"\0")]
    )


def benchmark_geojson(raw_data, verbose=True):
    for compression, module in tqdm(((None, None), ("gzip", gzip), ("bz2", bz2)), "GeoJSON", leave=verbose):
        yield ("GeoJSON", compression, *measure_geojson(raw_data, module))
```

```python
# HDF5
# https://pandas.pydata.org/pandas-docs/stable/user_guide/io.html#io-perf
def measure_hdf5(df, format=None, complib=None, complevel=0):
    return _measure(
        "tmp.h5",
        lambda f: df.to_hdf(f, 'test', mode='w', format=format, complib=complib, complevel=complevel),
        lambda f: pd.read_hdf(f, 'test')
    )
                    

hdf5_compressors = (
    'zlib', 'lzo', 'bzip2', 'blosc:blosclz', 'blosc:lz4', 'blosc:lz4hc',
    'blosc:snappy', 'blosc:zlib', 'blosc:zstd'
)

def benchmark_hdf5(df, verbose=True):
    for format_ in tqdm(("fixed", "table"), "HDF5", leave=verbose):
        try:
            yield (f"HDF5 ({format_})", None, *measure_hdf5(df, format_))
            for compressor in tqdm(hdf5_compressors, "Compression", leave=False):
                yield (f"HDF5 ({format_})", compressor, *measure_hdf5(df, format_, compressor, 9))
        except NotImplementedError as e:
            # fixed format can't store categoricals
            if verbose:
                print(f"{format_}: {e}")
```

```python
# Feather
def measure_feather(df):
    return _measure(
        "tmp.feather",
        lambda f: df.to_feather(f),
        pd.read_feather
    )


def benchmark_feather(df, verbose=None):
    yield ("Feather", None, *measure_feather(df))
```

```python
# Parquet
def measure_parquet(df, engine, compression=None):
    return _measure(
        "tmp.parquet",
        # Explicitly don't store index
        lambda f: df.to_parquet(f, engine=engine, compression=compression, index=False),
        lambda f: pd.read_parquet(f, engine=engine)
    )


def benchmark_parquet(df, verbose=True):
    for engine in tqdm(("PyArrow", "fastparquet"), "Parquet", leave=verbose):
        for compression in tqdm((None, 'snappy', 'gzip', 'brotli'), "Compression", leave=False):
            try:
                yield (f"Parquet ({engine})", compression, *measure_parquet(df, engine.lower(), compression))
            except RuntimeError as e:
                # fastparquet doesn't support Snappy
                if verbose:
                    print(f"{engine}, {compression}: {e}")
```

```python
# Pickle
def measure_pickle(df, compression):
    return _measure(
        "tmp.pickle",
        lambda f: df.to_pickle(f, compression=compression),
        lambda f: pd.read_pickle(f, compression=compression)
    )

def benchmark_pickle(df, verbose=True):
    for compression in tqdm((None, 'gzip', 'bz2', 'zip', 'xz'), "Pickle", leave=verbose):
        yield ("pickle", compression, *measure_pickle(df, compression))
```

```python
def benchmark_all(raw_data=None, df=None, verbose=True):
    # TODO: better interface
    assert df is not None
    
    # Feather can't store non-default indices
    assert type(df.index) is pd.RangeIndex
    
    return measurements_to_df(itertools.chain(
        benchmark_geojson(raw_data, verbose) if raw_data else [],
        benchmark_hdf5(df, verbose),
        benchmark_feather(df, verbose),
        benchmark_parquet(df, verbose),
        benchmark_pickle(df, verbose)
    ))
```

## Set up plots

```python
# https://stackoverflow.com/a/57211901
def get_text_bbox(ax, text):
    """Get the bounding box in data coords for a matplotlib Text or Annotation.
    
    Note: must have called fig.canvas.draw() first.
    Note: the bounding box depends on xlim and ylim, so if you call this and then
          add an extreme data point to the plot then the result will change.
    """
    return text.get_window_extent().inverse_transformed(ax.transData)


def draw_annotations(ax, y, label, value):
    value_x = value
    
    if label:
        t = ax.annotate(
            label,
            xy=(0, y),
            xytext=(3, 0),  # 3 points offset
            textcoords="offset points",
            ha="left", va="center",
        )
        
        # Move value over if too wide
        bbox = get_text_bbox(ax, t)
        offset_in_data_coords = bbox.xmin 
        if value - bbox.xmax < offset_in_data_coords:
            t.xy = (value, y)
            value_x = get_text_bbox(ax, t).xmax

    t = ax.annotate(
        f"{value:.1f}",
        xy=(value_x, y),
        xytext=(3, 0),  # 3 points offset
        textcoords="offset points",
        ha="left", va="center"
    )
    
    # Update axes xlim to fit annotation
    # https://stackoverflow.com/a/32637550
    bbox = get_text_bbox(ax, t)
    if bbox.xmax > ax.get_xlim()[1]:
#         print(label, value, ax.get_xlim()[1], bbox)
        ax.update_datalim(bbox.corners())
        ax.autoscale_view()


# https://matplotlib.org/3.1.1/gallery/lines_bars_and_markers/barchart.html
def draw_grouped_bars(ax, df, major_field, minor_field, value_field, bar_size=0.35, gap_size=0.1):
    groups = df.groupby(major_field, sort=False)
    bar_palette = sns.color_palette("muted", len(groups))
    error_palette = sns.color_palette("dark", len(groups))
    scatter_palette = sns.color_palette("pastel", len(groups))
    # Avoid extending data limits just for outliers (hack)
    scatter_limit = df.groupby([major_field, minor_field], sort=False)[value_field].quantile(0.75).max()
    # For y ticks
    labels = groups.groups.keys()
    label_positions = []
    # For bars
    last_group_edge = gap_size
    # For annotations
    annotations_to_draw = []
    
    # Draw all the bars
    for i, (_, group) in enumerate(groups):
        value_groups = group.groupby(minor_field, sort=False)
        
        group_size = len(value_groups)*bar_size
        label_positions.append(last_group_edge + group_size/2)
        
        y = last_group_edge + np.arange(len(value_groups))*bar_size + bar_size/2
        last_group_edge += group_size + gap_size
        bar_values = value_groups[value_field].median()
        bar_labels = value_groups.groups.keys()
        ax.barh(y, bar_values, bar_size, color=bar_palette[i])
        
        for yi, (_, scatter_values) in zip(y, value_groups[value_field]):
            # TODO: skip if range is smaller than some proportion of axis size
            if len(scatter_values.unique()) <= 1:
                continue
            ax.plot(
                scatter_values.quantile((0.25, 0.75)), (yi, yi), 
                c=error_palette[i], alpha=0.5, zorder=3
            )
            scatter_values = scatter_values[scatter_values <= scatter_limit]
            scatter_y = yi*np.ones(len(scatter_values))
            # TODO: marker size proportional to bar size
            scatter_y += np.random.uniform(-bar_size/4, bar_size/4, scatter_y.shape)
            ax.plot(
                scatter_values, scatter_y, 'o', ms=5,
                c=scatter_palette[i], alpha=0.2, mew=0, zorder=2
            )
                
        
        annotations_to_draw += list(zip(y, bar_labels, bar_values))
    
    # Draw annotations in order of value, so that (hopefully) the first
    # one updates the xlim and then it settles down, keeping the others
    # from getting messed up,
    annotations_to_draw.sort(key=lambda v: v[-1], reverse=True)
    plt.gcf().canvas.draw()  # for get_window_extent
    for yi, label, value in annotations_to_draw:
        draw_annotations(ax, yi, label, value)

    ax.set_yticks(label_positions)
    ax.set_yticklabels(labels)
    ax.set_ylim((last_group_edge, 0))
    ax.set_xlim(xmin=0)
```

### Commentary

* Compressing the original GeoJSON files rapidly becomes impractical.
* Although not the best at anything, HDF5 is not bad (depending on the compression algorithm) and quite flexible.
* Feather is quite fast, although not the most compact.
* With PyArrow, uncompressed Parquet is nearly as fast as Feather but quite a bit smaller. Compressed Parquet is even smaller and still reasonably fast.
* With fastparquet, uncompressed Parquet produces much larger files than PyArrow (for some reason), and gzip compression produces smaller files than PyArrow (for some reason).
* Pickle is still surprisingly competitive, produce file sizes comparable to Feather with no compression and compressed Parquet with zip compression (although a bit slower for both).

I'll use Parquet with PyArrow and gzip, which should produce TODO per day and take TODO to write.
Parquet has the advantage of being able to natively handle a folder of files as a single dataset.

# Check round-trip safety

```python
def assert_same(a, b):
    assert ((a == b) | (pd.isna(a))).all().all(), "Different values"
    assert (a.dtypes == b.dtypes).all(), "Different dtypes"

    
# HDF5 (fixed)
# Can't store categoricals
df = deduped_df[:1000]
df.to_hdf("tmp.h5", "temp", mode="w", format="fixed")
tmp = pd.read_hdf("tmp.h5", "temp")
assert_same(df, tmp)
os.remove("tmp.h5")

# Parquet (PyArrow)
# Categoricals read as primitive type
df.to_parquet("tmp.parquet", engine="pyarrow", index=True)
tmp = pd.read_parquet("tmp.parquet")
assert_same(df, tmp)
os.remove("tmp.parquet")

# Parquet (fastparquet)
# Categoricals read as primitive type
df.to_parquet("tmp.parquet", engine="fastparquet", compression="gzip")
tmp = pd.read_parquet("tmp.parquet")
assert_same(df, tmp)
os.remove("tmp.parquet")

# HDF5 (table)
df = categorical_df[:1000]
df.to_hdf("tmp.h5", "temp", mode="w", format="table")
tmp = pd.read_hdf("tmp.h5", "temp")
assert_same(df, tmp)
os.remove("tmp.h5")

# Feather
df.to_feather("tmp.feather")
tmp = pd.read_feather("tmp.feather")
assert_same(df, tmp)
os.remove("tmp.feather")

# Pickle
df.to_pickle("tmp.pickle")
tmp = pd.read_pickle("tmp.pickle")
assert_same(df, tmp)
os.remove("tmp.pickle")
```

# Performance trend with dataset size

```python
measurements_trend = []
for i in tqdm(np.unique(np.geomspace(1, 60, 15, dtype=int))):
    bulk_df = pd.concat(obs[:i]).reset_index(drop=True)
    measurements = benchmark_all(geojson_data[:i], bulk_df, verbose=False)
    measurements.insert(2, "number_of_obs", i)
    measurements_trend.append(measurements)
measurements_trend = pd.concat(measurements_trend)
measurements_trend.file_size = measurements_trend.file_size/1000
measurements_trend.groupby(["format", "compression", "number_of_obs"], sort=False).median()
```

```python
# One colour per format
formats = measurements_trend.format.unique()
colours = dict(zip(formats, sns.color_palette("bright", len(formats))))

# One line style per compression for each format
compressions = {
    format_: list(data.compression.unique())
    for (format_, data) in measurements_trend.groupby(["format"])
}
linestyles = ["-", "o-", "v-", "^-", "<-", ">-", "s-", "p-", "+-", "D-"]
styles = {format_: dict(zip(c, linestyles)) for (format_, c) in compressions.items()}


plt.figure(figsize=(12, 10))
for (format_, compression), df in measurements_trend.groupby(["format", "compression"], sort=False):
    tmp = df.groupby("number_of_obs").median().reset_index()
    x = tmp.number_of_obs
    y = tmp.file_size/tmp.number_of_obs
    style = styles[format_][compression]
    label = format_
    if compression:
        label += ", " + compression
    colour = colours[format_]
    
    # Line, without markers
    plt.plot(x, y, style, label=label, color=colour, lw=1, markevery=[])
    
    if len(style) > 1:
        # Manually place markers
        t = compressions[format_].index(compression)/len(compressions[format_])
        mx = np.arange(x.min() + t*7.5, x.max(), 10)
        my = np.interp(mx, x, y)
        plt.plot(mx, my, style[0], color=colour, alpha=0.5)
        
plt.xlabel("Number of observations per station")
plt.ylabel("Normalised file size (kB)")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5));
```
